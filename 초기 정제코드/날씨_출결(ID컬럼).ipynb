```python
import pandas as pd
import numpy as np
```

```python
w=pd.read_csv('정제데이터/날짜별 날씨 온도_Temp.csv' ,encoding='cp949')
w
```

```python
w.dtypes
```

```python
# '기온'열 값이 0 미만인 행: '날씨'열 값을 '한파'로 설정
w.loc[w['기온']<0,'날씨']='한파'
# 열 이름 변경
w=w.rename(columns={'날짜':'Date','기온':'Temp','날씨':'Weather'})

# obj-> datetime 변경
w['Date'] = pd.to_datetime(w['Date'])
w
```

```python
# 출결 org 불러오기
att=pd.read_excel('오리지널데이터/출결_통합본_org.xlsx')

# 일실, 퇴실 드랍
att=att.drop(columns=['입실시간','퇴실시간'])

# 날짜 형식 바꾸기
att['일자']=att['일자'].apply(lambda x: pd.to_datetime(str(x),format='%Y-%m-%d'))

# 열 이름 변경
att=att.rename(columns={'구분':'ID','일자':'Date','출결상태':'Stat'})
att
```

```python
# ID_1 출결사항 뽑기
ID_1=att[(att['ID'].astype('str')=='1')]

ID_1=ID_1.drop(columns='ID')

# ID_1의 df 열 이름 : 'Date','ID_1'
ID_1.columns=['Date','ID_1']

# datetime으로 변경
ID_1['Date'] = pd.to_datetime(ID_1['Date'])
ID_1
```

```python
# ID_1,w 합치기
w = pd.merge(w, ID_1)
w
```
