```python
import pandas as pd
import numpy as np
```

```python
c_df=pd.read_csv('오리지널데이터/출결_시간형식_통합본.csv',encoding='cp949')
c_df
```

```python

c_df = c_df[c_df['출결상태'] != '결석']

c_df
```

```python
# 입실시간 데이터 타입 시간 -> int로 변경 
c_df['입실시간'] = pd.to_datetime(c_df['입실시간'], format='%H:%M:%S').dt.strftime('%H%M').astype(int)
c_df['퇴실시간'] = pd.to_datetime(c_df['퇴실시간'], format='%H:%M:%S').dt.strftime('%H%M').astype(int)



c_df
```

```python
c_df = c_df.drop('Unnamed: 0', axis=1)
```

```python
c_df['일자'] = pd.to_datetime(c_df['일자'], format='%Y-%m-%d')
c_df
```

```python
d_df=pd.read_excel('오리지널데이터/교통카드이용내역_통합_org.xls')
d_df
```

```python
d_df['이용날짜'] = pd.to_datetime(d_df['이용날짜'].astype(str), format='%Y-%m-%d')
d_df['이용날짜'] = d_df['이용날짜'].dt.strftime('%Y-%m-%d')

d_df
```

```python
d_df['이용날짜'] = pd.to_datetime(d_df['이용날짜'])
d_df = d_df.rename(columns={'이용날짜':'일자'})
d_df.dtypes
```

```python
d_df
```

```python
# 이용시간 데이터 타입 변경  시간 -> int 
d_df['이용시간'] = pd.to_datetime(d_df['이용시간'], format='%H:%M').dt.strftime('%H%M').astype(int)
```

```python
d_df
```

```python
d_df = d_df[d_df['이용시간']<1300]

d_df
```

```python
a_df=pd.merge(c_df,d_df)
a_df
```

```python
a_df=a_df.drop(['퇴실시간','출결상태'],axis=1)
a_df
```

```python
a_df['reqTime'] = a_df['입실시간'] - a_df['이용시간']

a_df
```

```python
a_df=a_df.drop(['입실시간','이용시간'],axis=1)

a_df
```

```python
a_df.to_csv('정제데이터/평균소요시간.csv',encoding='cp949')
```
