```python
import pandas as pd
import numpy as np
import datetime
```

```python
weather=pd.read_excel('오리지널데이터/날씨정보_org.xlsx')
weather
```

```python
#강수량 null 값에 0 값넣기
weather["강수량"]=weather["강수량"].fillna(0)
weather
```

```python
# weather['날씨']=weather[(weather['강수량'].isna() != False) |(weather['강수량'] != 0),'비']
# weather
```

```python
#강수량이 0이면 맑음 0보다 크면 비 
weather['날씨']=np.where(weather['강수량']>0,'비','맑음')
weather
```

```python
# 날짜별 평균 기온값
dfaverage=weather.groupby('날짜')['기온'].mean()
dfaverage
```

```python
#날짜 병합
weatherID=weather['날짜'].unique()
weatherID
```

```python
weather
```

```python
#해당 날짜에 비가 하나라도 있으면 비 없으면 맑음
weatherlist=[]
for i in weatherID : 
  df=weather.loc[weather['날짜']==i,:] #weather 날짜와 i(병합한 날짜) 가 같으면 병합한날짜로 표시
  if (df['날씨']=='비').any():#날씨가 비가 하나라도(any()) 있으면
    weatherlist.append('비')# 비로 변경
  else:
    weatherlist.append('맑음')

weatherlist    
```

```python
#날짜별 날씨 온도 병합
TotalWeather=pd.DataFrame(dfaverage) #데이터 프레임 표시
TotalWeather['날씨']=weatherlist # 날씨에 weatherlist 배열값을 대입
TotalWeather.reset_index(inplace=True)# 컬럼 배열 정리
TotalWeather['기온'] = TotalWeather['기온'].apply(lambda x: round(x, 1))# 온도 0.1까지 표시
TotalWeather.to_csv('정제데이터/날짜별 날씨 온도_Temp.csv',index=False, encoding='cp949')
TotalWeather
```
