```python
import pandas as pd
import numpy as np
import datetime
```

```python
# 일자별 평균시간(정상) 파일 불러오기
reqTime=pd.read_csv('정제데이터/일자별 평균시간(정상).csv')
reqTime
```

```python
reqTime.info()
```

```python
# reqTime 일자 컬럼(object)을 datetime 객체로 변환
reqTime['일자_datetime']=pd.to_datetime(reqTime['일자'])
reqTime
```

```python
#필요 데이터만 추출
reqTime=reqTime[['일자_datetime','mean_reqTime']]
reqTime
```

```python
#열이름변경
reqTime.rename(columns = {'일자_datetime' : '일자'}, inplace = True)
reqTime
```

```python
#출석부 불러오기
df=pd.read_excel('오리지널데이터/출결_시간형식_통합본_org.xlsx')

#필요데이터 추출
df=df[['일자','출결상태']]
df
```

```python
# 지각한 날만 출력
df=df[(df['출결상태']=='지각') | (df['출결상태']=='지각&조퇴')]
df
```

```python
# 지각한날 날짜별 정렬
df=df.sort_values(by='일자')
df
```

```python
# 지각한날 중복일자 제거
df1=df.drop_duplicates(['일자'])
df1
```

```python
#출석부랑 reqtime 병합(지각 일자 기준)
result=pd.merge(reqTime, df1, on='일자')
result
```

```python
# 월별 출력
result['월'] = result['일자'].dt.month
result
```

```python
# 월별 평균 소요 시간(지각, 지각&조퇴 시)
result1=result.groupby('월')['mean_reqTime'].agg(**{'mean_reqTime':'mean'}).reset_index().round(2).reindex([2,3,4,5,6,0,1])
result1.to_csv("정제데이터/월별 평균소요시간(지각).csv",encoding='cp949')
```
