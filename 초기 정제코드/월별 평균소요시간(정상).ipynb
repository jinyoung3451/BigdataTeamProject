```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc, font_manager
# 한글 처리를 위해 폰트 설정
font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)
```

```python
#데이터 불러오기(결석 제외 평균 소요시간)
df=pd.read_excel("정제데이터/평균소요시간(정상_결측치보완).xlsx")
# df=df.drop(['ID'], axis=1)
# df['일자'] = pd.to_datetime(df['일자'],format='%Y-%m')
# reqTime 100분 이하 데이터만 출력
# df=df[df['reqTime'] <= 100.0]
df
```

```python
#ID 제거
df=df.drop(['ID'], axis=1)
df
```

```python
#일자별 평균 소요 시간(정상)
result1=df.groupby('일자')['reqTime'].agg(**{'mean_reqTime':'mean'}).reset_index()
result1
```

```python
#요일별 출력
df1=df
df1['요일'] = df1['일자'].dt.day_name()
df1
```

```python
#요일별 평균 소요 시간(정상)
result2=df1.groupby('요일')['reqTime'].agg(**{'mean_reqTime':'mean'}).reset_index().round(2).reindex([1,3,4,2,0])
result2
```

```python
# 월별 출력
result1['월'] = result1['일자'].dt.month
result1
```

```python
# 월별 평균 소요 시간 출력
result3=result1.groupby('월')['mean_reqTime'].agg(**{'mean_reqTime':'mean'}).reset_index().round(2).reindex([2,3,4,5,6,0,1])
result3
```

```python
# csv 파일로 저장
result3.to_csv('정제데이터/월별 평균소요시간(정상).csv', encoding='cp949', index=False)
```
