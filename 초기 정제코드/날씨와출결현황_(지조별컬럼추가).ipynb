```python
import pandas as pd
import numpy as np
```

```python
WA=pd.read_csv('정제데이터/날씨_출결(ID컬럼).csv',encoding='CP949')
WA
```

```python
# '지각'열 추가, 값 0으로 초기화
WA['지각']=0

# 각 행 반복
for i, row in WA.iterrows():
    
    # '지각' 발생 횟수 계산
    late_count=(row[4:8]=='지각').sum()

    # '지각' 열 값을 'late_count'로 지정
    WA.at[i, '지각']=late_count

WA
```

```python
# '조퇴'열 추가, 값 0으로 초기화
WA['조퇴']=0

# 각 행 반복
for i, row in WA.iterrows():
    
    #  '조퇴' 발생 횟수 계산
    early_count=(row[4:8]=='조퇴').sum()

    # 'Late' 열 값을 'early_count'로 지정
    WA.at[i, '조퇴']=early_count

WA
```

```python
# '결석'열 추가, 값 0으로 초기화
WA['결석']=0

# 각 행 반복
for i, row in WA.iterrows():
    
    # '결석' 발생 횟수 계산
    absent_count=(row[4:8]=='결석').sum()

    # '결석' 열 값을 'absent_count'로 지정
    WA.at[i, '결석']=absent_count

WA
```

```python
# '지각&조퇴'열 추가, 값 0으로 초기화
WA['지각&조퇴']=0

# 각 행 반복
for i, row in WA.iterrows():
    
    #  '지각&조퇴' 발생 횟수 계산
    late_ear_count=(row[4:8]=='지각&조퇴').sum()

    # '지각&조퇴' 열 값을 'late_ear_count'로 지정
    WA.at[i, '지각&조퇴']=late_ear_count

WA
```

```python
# '출석'열 추가, 값 0으로 초기화
WA['출석']=0

# 각 행 반복
for i, row in WA.iterrows():
    
    #  '출석' 발생 횟수 계산
    attend_count=(row[4:8]=='출석').sum()

    # '출석' 열 값을 'attend_count'로 지정
    WA.at[i, '출석']=attend_count

WA
```

```python
WA.to_csv('정제데이터/날씨_출결현황(지조별컬럼추가).csv',index=False,encoding='CP949')
```
