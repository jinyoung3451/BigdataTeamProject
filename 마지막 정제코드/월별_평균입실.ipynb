```python
import pandas as pd
import numpy as np
import datetime
```

```python
org_df=pd.read_excel('오리지널데이터/출결_시간형식_통합본_org.xlsx')
org_df
```

```python
org_df.rename(columns={'구분':'ID','일자':'Date','입실시간':'InTime','퇴실시간':'OutTime','출결상태':'Stat'},inplace=True)
org_df
```

```python
org_df['Date'] = org_df['Date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
```

```python
org_df
```

```python
#문자열을 datetime 객체로 변환
org_df['InTime'] = pd.to_datetime(org_df['InTime'],format='%H:%M:%S', errors='raise')

#시간 표기 변환
org_df['InTime']=org_df['InTime'].dt.time

org_df.dtypes
```

```python
org_df
```

```python
#시간 -> 정수형 변환
org_df['InTime']=org_df['InTime'].apply(lambda x: x.hour*3600 + x.minute*60 + x.second)
org_df
```

```python
# 입실시간 0인것은 제외
org_df=org_df[org_df.InTime != 0]
org_df
```

```python
# 일자별 평균 입실 시간 구하기
result=org_df.groupby('Date')['InTime'].agg(**{'평균입실시간':'mean'}).reset_index()

result['평균입실시간'] = result['평균입실시간'].astype(int)
result
```

```python
for i in result.index:
    in_time_str = str(result.loc[i, '평균입실시간']) # InTime 값을 문자열로 변환
    if ':' in in_time_str: # 이미 시간 형식으로 저장된 데이터인 경우
        in_time = datetime.datetime.strptime(in_time_str, '%H:%M:%S').time() # 시간 형식으로 변환
    else: # 정수형으로 저장된 데이터인 경우
        in_time_sec = int(in_time_str) # 문자열을 정수형으로 변환
        in_time = datetime.time(hour=in_time_sec//3600, minute=(in_time_sec%3600)//60, second=in_time_sec%60) # 시간 형식으로 변환
    result.loc[i, '평균입실시간'] = in_time
```

```python
result
```

```python
#InTime을 int로 변화
result['평균입실시간']=result['평균입실시간'].apply(lambda x:int(''.join(str(x).replace(':',''))))

result
```

```python
result.to_csv("완료데이터/월별평균입실.csv", encoding='cp949')
result
```
