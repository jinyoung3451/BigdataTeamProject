```python
import pandas as pd
import numpy as np
import datetime
```

```python
# 월별 평균소요시간(정상, 지각) 파일 불러오기
reqTime_normal=pd.read_csv('정제데이터/월별 평균소요시간(정상).csv', encoding='cp949')
reqTime_late=pd.read_csv('정제데이터/월별 평균소요시간(지각).csv', encoding='cp949')

# 열 이름 변경
reqTime_normal.rename(columns = {'mean_reqTime' : '정상'}, inplace=True)
reqTime_late.rename(columns = {'mean_reqTime' : '지각'}, inplace=True)
reqTime_late
```

```python
#월별 정상 지각 평균 소요 시간 합치기
reqTime_result=pd.merge(reqTime_normal, reqTime_late, on='월')
reqTime_result
```

```python
#월 컬럼 이름 변경
reqTime_result=reqTime_result.assign(월 = ['8월','9월','10월','11월','12월','1월','2월'])
reqTime_result
```

```python
reqTime_result.to_csv("정제데이터/월별_평균소요시간.csv",encoding='cp949')
```
