```python
import pandas as pd
```

```python
df=pd.read_excel("정제데이터/평균소요시간(정상_결측치보완).xlsx")
df
```

```python
df=pd.read_excel('오리지널데이터/출결_시간형식_통합본_org.xlsx')
df
```

```python
#지각만 나열
df = df[df['출결상태'] == '지각']
df
```

```python
#구분 퇴실시간 제거
df = df.drop(['구분','퇴실시간'],axis=1)
```

```python
df
```

```python
#요일컬럼 추가
df["요일"] = df["일자"].dt.day_name()
```

```python
#일자별로 정렬후 인덱스 새로주기
df=df.sort_values(['일자']).reset_index()
```

```python
#기존인덱스 제거
df=df.drop(['index'],axis=1)
df
```

```python
df2 = pd.read_csv('정제데이터/소요시간_temp.csv')
df2
```

```python
#컬럼이름 재정비 후 같은날자로 뭉개고 평균소요시간 계산
df2.rename(columns={'Date':'일자'}, inplace=True)
df2=df2.drop(['ID'],axis=1)
df2=df2.groupby('일자')['reqTime'].agg(**{'mean_reqTime':'mean'}).reset_index()
df2
```

```python
df2['일자'] = pd.to_datetime(df2['일자']) # 일자 컬럼을 datetime 형식으로 변환
df2.dtypes
```

```python
#합치기
df3=pd.merge(df,df2,on='일자')
df3
```

```python
# 소요시간 정수형으로 전환
df3['mean_reqTime'] = df3['mean_reqTime'].round().astype(int)

df3
```

```python
# 각 요일별 평균 소요시간 값을 담을 리스트 초기화
mean_reqTime_list = [0] * 5

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# 요일별 평균 소요시간 계산 후 리스트에 추가
for i in range(5):
    mean_reqTime_list[i] = df3[df3['요일'] == days[i]]['mean_reqTime'].mean()

print(mean_reqTime_list)
```

```python

#소요시간 소숫점2자리로 정리
mean_reqTime_list = [round(x, 2) for x in mean_reqTime_list]
mean_reqTime_list
```

```python
#비정상과 비교하기 위해 정상출결 평균시간 데이터 불러오기
nomal_df=pd.read_csv('data/요일별 평균시간(정상).csv')
nomal_df
```

```python
#정상출결일때 평균 소요시간 추출
mean_nomal = nomal_df['mean_reqTime'].values
mean_nomal
```

```python
#요일별 정상 지각 평균시간 데이터프레임 만들기

all_df=pd.DataFrame({
    '요일': days,
    '정상':mean_nomal,
    '지각': mean_reqTime_list
    
    }
)

all_df
```

```python
#저장

all_df.to_csv('완료데이터/요일별_평균소요시간.csv',encoding='cp949')
```

```python
#비교분석을 위해 날씨 출결데이터 불러오기

w_df=pd.read_csv('정제데이터/날씨와출결현황.csv', encoding='cp949')
w_df
```

```python
#날씨 출결 데이터에서 비정상 출결사항 추출후 배열에 담기
weather_chul = w_df['비&한파 합계'].values
weather_chul
```

```python
#지각일 경우 교통시간이 정상 교통시간보다 많은 경우의 수를 담기
gyotong_chul=[]
```

```python
#월,화는 해당사항없고 수요일 부터 정상평균보다 교통소요시간이 늦은 경우 추출

count = len(df3[(df3['요일'] == 'Wednesday') & (df3['mean_reqTime'] >= 38.31)])

gyotong_chul.append(count)

gyotong_chul
```

```python
#월,화는 해당사항없고 수요일 부터 정상평균보다 교통소요시간이 늦은 경우 추출

count = len(df3[(df3['요일'] == 'Thursday') & (df3['mean_reqTime'] >= 37.42)])

gyotong_chul.append(count)
```

```python
#월,화는 해당사항없고 수요일 부터 정상평균보다 교통소요시간이 늦은 경우 추출

count = len(df3[(df3['요일'] == 'Friday') & (df3['mean_reqTime'] >= 38.58)])

gyotong_chul.append(count)

gyotong_chul
```

```python
#위의 경우를 전부 더하고 교통요인으로 인해 조퇴,결석,지각&조퇴 경우는 없으므로 0으로 대체후 배열만들기
total_gyotong_chul=sum(gyotong_chul)

gyotong_chul_array=[total_gyotong_chul,0,0,0]

gyotong_chul_array
```

```python
gubun=['지각','조퇴','결석','지각&조퇴']
```

```python
#기타요인을 계산하기 위해 임시로 데이터프레임 작성

all_df2=pd.DataFrame({
    '구분': gubun,
    '날씨':weather_chul,
    '교통': gyotong_chul_array
    
    }
)

all_df2
```

```python
#기타 요인을 계산하기 위해 월별 출결데이터를 불러옴

m_df=pd.read_csv('완료데이터/월별출결.csv',encoding='cp949')
m_df
```

```python
#지각 조퇴 결석 지각&조퇴 순으로 맞추기 위해 컬럼재정렬

m_df = m_df[['월', '출석', '지각', '조퇴', '결석', '지각&조퇴', '출결총합', '출석률']]



m_df
```

```python
#총합만 구하기 위해 7번째 행에서 2번열부터 6번째 열 값 추출해서 배열에 담기

chong = m_df.iloc[7, 2:6].values.tolist()
chong
```

```python
#기타 요인을 계산하기 위해 총합 - 나머지 요인들을 하기 위해 다시 임시로 데이터프레임 작성

all_df2=pd.DataFrame({
    
    '구분': gubun,
    '총합':chong,
    '날씨':weather_chul,
    '교통': gyotong_chul_array
    
    }
)

all_df2
```

```python
#기타요인 계산후 컬럼추가

all_df2['기타'] = all_df2['총합'] - (all_df2['날씨'] + all_df2['교통'])

all_df2
```

```python
#저장
all_df2.to_csv('완료데이터/비교분석데이터.csv', encoding='cp949')
```
