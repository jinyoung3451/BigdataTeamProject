```python
import pandas as pd
import numpy as np
```

```python
#데이터 불러오기
df=pd.read_csv('완료데이터/월별출결.csv',encoding='cp949')

df
```

```python
#필요 없는컬럼 제거
df=df.drop(columns=['Unnamed: 0'],axis=1)
df
```

```python
#인덱스 7번쨰 총합부분 제거
df=df.drop(7)
df
```

```python
# 출석률 리스트 가져오기
per_df= df['출석률'].values.tolist()

per_df
```

```python
#출석률 변화 계산값
per = []
for i in range(len(per_df)-1): # 여기서 -1은 end 라는 뜻
    diff = per_df[i+1] - per_df[i] # ex ) 9월 - 8월 
    per.append(diff)

per
```

```python
months=['9월','10월','11월','12월','1월','2월']
```

```python
all_df=pd.DataFrame({
    
    '월별': months,
    '출석률변화':per
   
    
    }
)

all_df
```

```python
all_df.to_csv('완료데이터/출석률변화그래프.csv',encoding='cp949')
```
